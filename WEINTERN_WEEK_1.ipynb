```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
sns.set(style="whitegrid")
```

```python
df['Age'].value_counts().plot(kind='bar')
plt.title("Bar Chart")
plt.xlabel("age")
plt.ylabel("number of people")
plt.show()
```

```python
plt.scatter(df['Department'], df['CGPA'])
plt.title("scatter plot")
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.show()
```

```python
sns.boxplot(x=df['Sleep_Duration'])
plt.title("Box Plot")
plt.show()
```

```python
plt.hist(df['Stress_Level'], bins=10)
plt.title("Histogram")
plt.xlabel("Values")
plt.ylabel("Frequency")
plt.show()
```

```python
sns.countplot(x='Department', data=df)
plt.show()
```

```python
sns.pairplot(df)
plt.show()
```

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df_encoded = pd.get_dummies(df, drop_first=True)
```

```python
plt.figure(figsize=(10,8))
sns.heatmap(df_encoded.corr(), annot=False, cmap="coolwarm")
plt.title("Correlation Heatmap (Encoded Data)")
plt.show()
```

2.The scatter plot illustrates the relationship between students departments and their CGPA values. Each point on the graph represents an individual student, with the department shown along the horizontal axis and the corresponding CGPA on the vertical axis. This visualization helps in understanding how CGPA values are distributed across different departments and makes it easier to observe variations, patterns, or overlaps in academic performance between departments. 3.The box plot illustrates the distribution of sleep duration values in the dataset and provides a clear summary of their spread and central tendency. It shows the median sleep duration as a line within the box, while the box itself represents the range between the first and third quartiles, capturing the middle 50% of the data. The whiskers extend to indicate the typical minimum and maximum values, excluding extreme cases, and any points beyond the whiskers represent outliers. This visualization is useful for identifying variations in sleep patterns, detecting unusually high or low sleep durations, and understanding the overall consistency of sleep among individuals in the dataset.

4.The  histogram illustrates the distribution of stress levels among the individuals in the dataset. The horizontal axis represents the range of stress level values, which are grouped into equal intervals, while the vertical axis shows the frequency of individuals falling within each range. By visualizing the data in this form, it becomes easy to understand how stress levels are spread across the population, identify commonly occurring stress ranges, and observe any patterns such as clustering, skewness, or extreme values. This visualization helps in gaining an overall understanding of stress intensity and variability within the dataset.

5.The count plot represents the distribution of employees across different departments in the dataset. Each bar corresponds to a department, and the height of the bar indicates the number of individuals belonging to that department. By displaying the frequency of records for each department, the chart makes it easy to compare department-wise representation at a glance. This helps in understanding how the workforce is distributed, identifying departments with higher or lower staff strength, and supporting data-driven decisions related to resource allocation and organizational planning.

6.The visualization of pair plot presents a pairwise comparison of all variables in the dataset, allowing relationships between different features to be examined simultaneously. Each row and column correspond to a variable, with the diagonal plots showing the distribution of individual variables and the off-diagonal plots illustrating how one variable relates to another. By viewing these patterns together, it becomes easier to identify trends, correlations, clusters, and potential outliers within the data. Overall, this visualization provides a comprehensive overview of the datasets structure and helps in understanding how multiple variables interact with one another.

7.The visualization of heatmap presents a correlation heatmap of the dataset after converting categorical variables into numerical form, allowing relationships between all features to be analyzed together. The heatmap illustrates how strongly different variables are related to one another, with color intensity representing the direction and strength of correlationwarmer colors indicate positive relationships, cooler colors indicate negative relationships, and neutral shades suggest little or no correlation. By excluding individual numeric labels and focusing on color patterns, the heatmap provides a clear, high-level view of how variables interact, helping to identify strongly related features, potential redundancies, and patterns that may be useful for data analysis and predictive modeling.
