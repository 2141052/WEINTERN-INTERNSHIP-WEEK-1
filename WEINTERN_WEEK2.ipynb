```python
from google.colab import files
uploaded = files.upload()

import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

df = pd.read_csv("Sales Dataset.csv")
df.head()
```

```python
df.groupby("Category")["Amount"].mean().plot(kind="bar")
plt.title("Average Amount by Category")
plt.xlabel("Category")
plt.ylabel("Average Amount")
plt.show()
```

```python
df.groupby("Category")["Amount"].mean().plot(kind="bar")

plt.title("Average Amount by Category")
plt.xlabel("Category")
plt.ylabel("Average Amount")
plt.show()
```

```python
plt.hist(df["Quantity"], bins=10)

plt.title("Quantity Distribution")
plt.xlabel("Quantity")
plt.ylabel("Frequency")
plt.show()
```

```python
df["Category"].value_counts().plot(kind="pie", autopct="%1.1f%%")

plt.title("people per Category")
plt.ylabel("")  # Removes extra label
plt.show()
```

```python
plt.scatter(df["Quantity"], df["Amount"])

plt.title("Quantity vs Amount")
plt.xlabel("Quantity")
plt.ylabel("Amount")
plt.show()
```

```python
sns.boxplot(x=df["Amount"])

plt.title("Boxplot of Amount")
plt.show()
```
